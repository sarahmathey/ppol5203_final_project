```python
import pandas as pd
import numpy as np
import re 
import os
import glob
```

```python
#os.chdir('~/Downloads')
csv_files = glob.glob('/Users/sarahmathey/Downloads/IHME_USA_COD_COUNTY_RACE_ETHN_2000_2019_MX_MATER_NEONAT_FEMALE/*.CSV')

# Create an empty list to store the dataframes
df_list = []

# Iterate through the CSV files, read them into dataframes, and append to the list
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    df_list.append(df)

# Concatenate all dataframes into one
final_df = pd.concat(df_list, ignore_index=True)
```

```python
# limit data to only counties 
county_data = final_df[final_df.location_name.str.contains('County')].copy()
```

```python
# filter data so that we're only pullint all races, all ages, and 2019
filter_df_total = county_data[county_data['race_name']=='Total'].copy()
filter_df_total = filter_df_total[filter_df_total['age_name']=='All Ages']
filter_df_total = filter_df_total[filter_df_total['year']==2019]
```
